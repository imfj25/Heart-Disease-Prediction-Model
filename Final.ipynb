```python
from tkinter import *
import joblib
import numpy as np
from sklearn import *
from tkinter import messagebox
master = Tk()
def entry():
    try:
        p1=int(e1.get())
        p2=int(e2.get())
        p3=int(e3.get())
        p4=int(e4.get())
        p5=int(e5.get())
        p6=int(e6.get())
        p7=int(e7.get())
        p8=int(e8.get())
        p9=int(e9.get())
        p10=float(e10.get())
        p11=int(e11.get())
        p12=int(e12.get())
        p13=int(e13.get())
    except:
        messagebox.showerror('Empty Fields','All Fields Are Mandatory !',icon = 'warning')
        
    model = joblib.load('model_joblib_heart')
    result=model.predict([[p1,p2,p3,p4,p5,p6,p7,p8,p8,p10,p11,p12,p13]])
    
    
    if result == 0:
        Label(master, text="No Heart Disease",fg="black",
              bg="#0096DC",font=("Impact",25,"bold")).place(x=750,y=480) 
    else:
        Label(master, text="Possibility of Heart Disease",fg="black",
              bg="#0096DC",font=("Impact",25,"bold")).place(x=750,y=480) 



def ExitApplication():
    
    box= messagebox.askquestion ('Exit Application','Are you sure you want to exit the application ?',
                                 icon = 'warning')
    
    if box== 'yes':    
        master.destroy()    
        
exit_button = Button(master, text="EXIT",fg="red",bd = '7',bg="white",
                     activebackground="red",font=("Classy Button",10),cursor="hand2",command=ExitApplication)
exit_button.place(x=1125,y=580)

    
    
master_width = 1200
master_height= 650
master.geometry(f"{master_width}x{master_height}+{30}+{0}")
master.config(bg="#0096DC")
master.title("Heart Disease Prediction")
master.iconbitmap('heartIcon.ico')
label = Label(master,text="H E A R T   D I S E A S E   P R E D I C T I O N",fg="green2",
              bg="#0096DC",font=("aesthetica",35,"bold"), borderwidth=1, relief="solid").place(x=180,y=20)


#----------------Age-----------------------
Label(master, text="1. Enter Your Age",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=130)
e1 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e1.place(x=410,y=130,height=23)

#----------------Gender-----------------------
Label(master, text="2. Male Or Female [1/0]",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=150,y=185)
e2 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e2.place(x=410,y=185,height=23)

#------------------CP---------------------
Label(master, text="3. Enter Value of CP",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=240)
e3 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e3.place(x=410,y=240,height=23)

#------------------trestbps---------------------
Label(master, text="4. Enter Value of trestbps",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=295)
e4 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e4.place(x=410,y=295,height=23)

#------------------chol---------------------
Label(master, text="5. Enter Value of chol",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=150,y=350)
e5 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e5.place(x=410,y=350,height=23)

#------------------fbs---------------------
Label(master, text="6. Enter Value of fbs",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=405)
e6 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e6.place(x=410,y=405,height=23)

#------------------restecg---------------------
Label(master, text="7. Enter Value of restecg",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=460)
e7 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e7.place(x=410,y=460,height=23)

#-------------------thalach--------------------
Label(master, text="8. Enter Value of thalach",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=150,y=515)
e8 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e8.place(x=410,y=515,height=23)

#-------------------exang--------------------
Label(master, text="9. Enter Value of exang",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=150,y=570)
e9 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e9.place(x=410,y=570,height=23)

#-------------------oldpeak--------------------
Label(master, text="10. Enter Value of oldpeak",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=690,y=130)
e10 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e10.place(x=950,y=130,height=23)

#-------------------slope--------------------
Label(master, text="11. Enter Value of slope",fg="black",bg="#0096DC", 
      font=("Helvetica",15,"bold")).place(x=690,y=185)
e11 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e11.place(x=950,y=185,height=23)

#-------------------ca--------------------
Label(master, text="12. Enter Value of ca",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=690,y=240)
e12 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e12.place(x=950,y=240,height=23)

#-------------------thal--------------------
Label(master, text="13. Enter Value of thal",fg="black",bg="#0096DC",
      font=("Helvetica",15,"bold")).place(x=690,y=295)
e13 = Entry(master,cursor="hand2", borderwidth=1, relief="solid")
e13.place(x=950,y=295,height=23)






Button(master,text='Submit',command=entry,bd = '7',activebackground='green',fg="green",bg="white",
       font=("Arial",10),cursor="hand2").place(x=1050,y=580)

mainloop()
```
